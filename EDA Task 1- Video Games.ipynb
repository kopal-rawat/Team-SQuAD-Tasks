# Exploratory Data Analysis And Visualization

## 1. Global Sales vs Rank

```python
plot1= df[0:25]
plot1
```

```python
x= plot1['Rank']
y= plot1['Year']
plt.plot(x,y,label='Year',color='red')
plt.ylabel('Year')
plt.xlabel('Rank')
plt.title('Relation between global sales by rank for first 25 rows')
plt.show()
```

## Overall Global Sales using Seaborn

```python
plt.figure(figsize=(25,8),dpi=80)
sns.kdeplot(df.Global_Sales, shade=True, label = 'Global Sales', color="r")
plt.title('Overall Global Sales Distribution')
plt.show()
```

## Sales Across North America, Europe And Japan

```python
NA=plot1['NA_Sales']
EUR=plot1['EU_Sales']
JAP=plot1['JP_Sales']
```

```python
plt.title('Comparision of Sales across the Countries')
plt.plot(NA, label = 'NORTH AMERICA')
plt.plot(EUR, label = 'EUROPE')
plt.plot(JAP, label = 'JAPAN')
plt.legend()
plt.show()
```

**North America has the most number of sales.**

## Most producing gaming platform

```python
plt.figure(figsize=(25,8),dpi=80)
plt.plot(df["Platform"].value_counts().index, df["Platform"].value_counts())
plt.title("Number of Games produced by a Gaming Platform")
plt.show()
```

**From the above graph it is visible that DS and PS2 produced the most number of games followed by PS3 and Wii.**

## Most popular Genre

```python
plt.figure(figsize=(25,8),dpi=80)
plt.plot(df[["Genre","Global_Sales"]].groupby("Genre").sum())
plt.title("Number of Games produced by a Gaming Platform")
plt.show()
```

**Action is the most sold genre followed by sports**
